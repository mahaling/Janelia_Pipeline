```python
import os,sys,json
```

```python
jsonfile = "generateTilePairsTemplate.json"

with open(jsonfile) as js:
    params = json.load(js)
    
zfirst = 2366
zlast = 4491
inc = 25
m = zfirst

for i in xrange(zfirst, zlast, inc):
    if i == 4466:
        j = i + inc
    else:
        j = i + inc - 1
    print i, j
    params["minZ"] = i
    params["maxZ"] = j
    if j == 4491:
        params["stack"] = "Phase1Data_" + str(m) + "_" + str(j)
    else:
        params["stack"] = "Phase1Data_" + str(m) + "_" + str(m+49)
    if (m+49 == j):
        m = m + 50
    print params["stack"]
    with open("tempJson.json", "w") as f:
        json.dump(params, f)
    cmd = "python 01_generate_tile_pairs.py --jsonfile tempJson.json"
    os.system(cmd)
    
```
